#### 6. Fix Incorrect Data Types
Based on what you discovered above, use appropriate methods to re-type incorrectly typed data.
- Define a function that will allow you to convert participation rates to an appropriate numeric type. Use `map` or `apply` to change these columns in each dataframe.

````python
#converting participation rates in df_sat2017 from object to float 

def object_to_float (every_series):
    """use .replace to remove '$' then convert to float"""
    return float(every_series.replace('%',''))/100       

#use.map to change participation rates in date frame df_sat2017

df_sat2017['Participation'] = df_sat2017['Participation'].map(object_to_float) 
````

````python
#To ensure the participation rates has been converted to float in in date frame df_sat2017
df_sat2017.head()
````

````python
#use.map to change participation rates in date frame df_sat2017
df_act2017['Participation'] = df_act2017['Participation'].map(object_to_float) 

````

````python
#To ensure the participation rates in date frame df_sat2017 has been converted to float
df_act2017.head()
````

- Fix any individual values preventing other columns from being the appropriate type.

````python
#As fixed the error in Q4, converting the data type of composite in df_act2017 from object to float 
df_act2017['Composite']= df_act2017['Composite'].astype(float)
````

- Finish your data modifications by making sure the columns are now typed appropriately.

````python
#Realize that still have 3 columns data types in df_sat2017 are integer. it 'd be better change to float

df_sat2017['Evidence-Based Reading and Writing']= df_sat2017['Evidence-Based Reading and Writing'].astype(float)
df_sat2017['Math']= df_sat2017['Math'].astype(float)
df_sat2017['Total']= df_sat2017['Total'].astype(float)

````
