#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#expression of the formular in python 
#std = (sum((x - mean)**2))/n)**0.5


def std_dev(datapoint): 
    """to find out the mean,the sum of distance of datapoint from the mean and square root """
    # Step 1: find the mean of the datapoint
    mean_x = np.mean(datapoint) 
    
    # Step 2 Find the square of the distance from every data point to the mean
    # Step 3 Sum of the distance and then divided by the number of datapoint
    in_root = sum([(i-mean_x)**2 for i in datapoint]) / len(datapoint)   
    
    #take the square root of step 3
    return in_root**0.5

````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms.
````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabel):#title):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2,figsize=(15,15)) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column]) # feel free to add more settings
        # Set titles, labels, etc here for each subplot
        ax[i].set_title(list_of_titles[i],fontsize=15, fontweight=0, color='red', loc='center')
        ax[i].set_xlabel(list_of_xlabels, fontsize=15)
        ax[i].set_ylabel("Number of States", fontsize=15)

plt.tight_layout()
plt.show()
````
