## Data Cleaning :  Drop columns with high percentage of null values, 
- because if marjority is in same catagory , not neccessary adding value. Take Pool QC as an example,the null values actually indicated that most of the house doesn't have a Pool. Though having a pool might affect saleprice but since majority of the house doesn't have a pool,the differenece will not be obvious. 
- In this case, dropping Pool Area and Fireplace as well

```python
#drop 9 columns 
#'PID' is the reference no. for the house, not contributing 

drop_cols = ['Pool QC', 'Misc Feature','Misc Val','Alley', 'Fence', 'Fireplace Qu','Pool Area', 'Fireplaces','PID']
df_train = df_train.drop(drop_cols, axis=1)
df_test = df_test.drop(drop_cols, axis=1)
```

## Data cleaning :handle some specific columns with mean and mode

```python
#LotFrontage missing value fill with train's mean as it is continous 

df_train['Lot Frontage']= df_train['Lot Frontage'].fillna(df_train['Lot Frontage'].mean())
```

```python
df_test['Lot Frontage' ]= df_test['Lot Frontage'].fillna(df_test['Lot Frontage'].mean())
```

```python
#only one missing value in test['Electrical'], fill with train's mode
df_test['Electrical' ]= df_test['Electrical'].fillna(df_test['Electrical'].mode()[0])
```
